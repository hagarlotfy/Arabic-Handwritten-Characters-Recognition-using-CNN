```python
import os
import pandas as pd

repo = "/content/Arabic-Handwritten-Characters-Recognition-using-CNN"
base = repo + "/my_dataset"

with open(base + "/classes.txt", "r", encoding="utf-8") as f:
    classes = [line.strip() for line in f.readlines()]

name_to_arabic = {
    "alif": "ا",
    "alif_hamza": "أ",
    "beh": "ب",
    "teh": "ت",
    "theh": "ث",
    "jeem": "ج",
    "hah": "ح",
    "khah": "خ",
    "dal": "د",
    "thal": "ذ",
    "zal": "ذ",
    "raa": "ر",
    "zain": "ز",
    "seen": "س",
    "sheen": "ش",
    "sad": "ص",
    "dad": "ض",
    "tah": "ط",
    "zah": "ظ",
    "ain": "ع",
    "ghain": "غ",
    "feh": "ف",
    "qaf": "ق",
    "kaf": "ك",
    "lam": "ل",
    "meem": "م",
    "noon": "ن",
    "heh": "ه",
    "waw": "و",
    "yaa": "ي",
    "teh_marbota": "ة",
    "hamza": "ء",
    "lam_alif": "لا",
}

def class_to_char(class_name):
    if class_name in name_to_arabic:
        return name_to_arabic[class_name]

    for suffix in ["_begin", "_middle", "_end", "_regular"]:
        if class_name.endswith(suffix):
            base_name = class_name.replace(suffix, "")
            return name_to_arabic.get(base_name, "?")

    return "?"

rows = []

for split in ["train", "val"]:
    img_dir = os.path.join(base, split, "images")
    label_dir = os.path.join(base, split, "labels")

    for txt_name in os.listdir(label_dir):
        if not txt_name.endswith(".txt"):
            continue

        txt_path = os.path.join(label_dir, txt_name)
        stem = os.path.splitext(txt_name)[0]

        image_file = None
        for ext in [".jpg", ".jpeg", ".png"]:
            candidate = stem + ext
            if os.path.exists(os.path.join(img_dir, candidate)):
                image_file = os.path.join(split, "images", candidate)
                break

        if image_file is None:
            continue

        boxes = []
        with open(txt_path, "r", encoding="utf-8") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) < 5:
                    continue

                class_id = int(parts[0])
                x_center = float(parts[1])
                class_name = classes[class_id]
                char = class_to_char(class_name)

                boxes.append((x_center, char, class_name))

        boxes = sorted(boxes, key=lambda x: x[0], reverse=True)
        word = "".join(char for _, char, _ in boxes)

        rows.append({
            "filename": image_file,
            "label": word,
            "num_letters": len(boxes),
            "classes": " ".join(c for _, _, c in boxes)
        })

df_words = pd.DataFrame(rows)
df_words.to_csv("/content/word_labels_from_yolo.csv", index=False, encoding="utf-8-sig")

print("created rows:", len(df_words))
df_words.head(20)
```

```python
import re
import pandas as pd

def clean_arabic_text(text):
    text = str(text)

    # normalize spaces
    text = text.replace("\xa0", " ")
    text = re.sub(r"\s+", " ", text).strip()

    # remove tatweel and diacritics
    diacritics = "ًٌٍَُِّْـ"
    for ch in diacritics:
        text = text.replace(ch, "")

    # remove weird Latin/control symbols
    allowed = set(" ابتثجحخدذرزسشصضطظعغفقكلمنهويىءآأؤإئة0123456789.,،؛:؟?!()-/\"")
    text = "".join(ch for ch in text if ch in allowed)

    return text.strip()

for path in ["/content/ocr_train.csv", "/content/ocr_val.csv"]:
    df = pd.read_csv(path)
    df["text"] = df["text"].apply(clean_arabic_text)
    df = df[df["text"].str.len() > 0].reset_index(drop=True)
    out = path.replace(".csv", "_clean.csv")
    df.to_csv(out, index=False, encoding="utf-8-sig")
    print(out, len(df))

train_df = pd.read_csv("/content/ocr_train_clean.csv")
val_df = pd.read_csv("/content/ocr_val_clean.csv")

all_text = "".join(train_df["text"].astype(str).tolist() + val_df["text"].astype(str).tolist())
chars = sorted(set(all_text))

char_to_idx = {ch: i + 1 for i, ch in enumerate(chars)}
idx_to_char = {i + 1: ch for i, ch in enumerate(chars)}

print("num chars:", len(chars))
print(chars)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
import numpy as np
import json
import os

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)
```

```python
IMG_H = 64
IMG_W = 512

with open("/content/arabic_ocr_vocab_clean.json", "r", encoding="utf-8") as f:
    vocab = json.load(f)

chars = vocab["chars"]
char_to_idx = vocab["char_to_idx"]
idx_to_char = {int(k): v for k, v in vocab["idx_to_char"].items()}

blank_idx = 0
num_classes = len(chars) + 1

def encode_text(text):
    return [char_to_idx[ch] for ch in text if ch in char_to_idx]

def resize_keep_ratio(img, target_h=IMG_H, target_w=IMG_W):
    img = img.convert("L")

    w, h = img.size
    new_w = int(w * (target_h / h))
    new_w = max(1, min(new_w, target_w))

    img = img.resize((new_w, target_h))

    canvas = Image.new("L", (target_w, target_h), color=255)
    canvas.paste(img, (0, 0))

    arr = np.array(canvas).astype(np.float32) / 255.0
    arr = 1.0 - arr  # black ink becomes high value

    return torch.tensor(arr).unsqueeze(0)

class ArabicOCRDataset(Dataset):
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["image_path"])
        x = resize_keep_ratio(img)

        y = torch.tensor(encode_text(str(row["text"])), dtype=torch.long)

        return x, y, str(row["text"])

def collate_fn(batch):
    images, labels, texts = zip(*batch)

    images = torch.stack(images)

    label_lengths = torch.tensor([len(l) for l in labels], dtype=torch.long)
    labels = torch.cat(labels)

    return images, labels, label_lengths, texts

train_ds = ArabicOCRDataset("/content/ocr_train_clean.csv")
val_ds = ArabicOCRDataset("/content/ocr_val_clean.csv")

train_loader = DataLoader(train_ds, batch_size=8, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_ds, batch_size=8, shuffle=False, collate_fn=collate_fn)

x, y, label_lengths, texts = next(iter(train_loader))

print("image batch:", x.shape)
print("labels:", y.shape)
print("label lengths:", label_lengths[:5])
print("sample text:", texts[0])
print("num_classes:", num_classes)
```

```python
class CRNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),   # 64x512 -> 32x256

            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),   # 32x256 -> 16x128

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)), # 16x128 -> 8x128

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)), # 8x128 -> 4x128
        )

        self.rnn = nn.LSTM(
            input_size=512 * 4,
            hidden_size=256,
            num_layers=2,
            bidirectional=True,
            batch_first=True,
            dropout=0.2
        )

        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        feat = self.cnn(x)          # B, C, H, W
        b, c, h, w = feat.size()

        feat = feat.permute(0, 3, 1, 2)  # B, W, C, H
        feat = feat.reshape(b, w, c * h) # B, T, features

        out, _ = self.rnn(feat)
        out = self.fc(out)              # B, T, classes

        return out

model = CRNN(num_classes).to(device)

with torch.no_grad():
    logits = model(x.to(device))

print("logits:", logits.shape)
```

```python
def decode_ctc(logits):
    # logits: B, T, C
    pred = logits.argmax(dim=-1).detach().cpu().numpy()

    texts = []

    for seq in pred:
        chars_out = []
        prev = None

        for idx in seq:
            if idx != blank_idx and idx != prev:
                chars_out.append(idx_to_char.get(int(idx), ""))
            prev = idx

        texts.append("".join(chars_out))

    return texts

criterion = nn.CTCLoss(blank=blank_idx, zero_infinity=True)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

def train_one_epoch():
    model.train()
    total_loss = 0

    for images, labels, label_lengths, texts in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        logits = model(images)  # B,T,C
        log_probs = F.log_softmax(logits, dim=-1).permute(1, 0, 2)  # T,B,C

        batch_size = images.size(0)
        input_lengths = torch.full(
            size=(batch_size,),
            fill_value=logits.size(1),
            dtype=torch.long
        ).to(device)

        label_lengths = label_lengths.to(device)

        loss = criterion(log_probs, labels, input_lengths, label_lengths)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5)
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)

def preview_predictions(n_batches=1):
    model.eval()

    with torch.no_grad():
        for images, labels, label_lengths, texts in val_loader:
            images = images.to(device)
            logits = model(images)
            preds = decode_ctc(logits)

            for true, pred in list(zip(texts, preds))[:5]:
                print("TRUE:", true)
                print("PRED:", pred)
                print("-" * 30)

            break

print("before training:")
preview_predictions()

loss = train_one_epoch()

print("epoch 1 loss:", loss)
print("after training:")
preview_predictions()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
import numpy as np
import json
import os

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else "NO GPU")

IMG_H = 64
IMG_W = 512

with open("/content/arabic_ocr_vocab_clean.json", "r", encoding="utf-8") as f:
    vocab = json.load(f)

chars = vocab["chars"]
char_to_idx = vocab["char_to_idx"]
idx_to_char = {int(k): v for k, v in vocab["idx_to_char"].items()}

blank_idx = 0
num_classes = len(chars) + 1

def encode_text(text):
    return [char_to_idx[ch] for ch in str(text) if ch in char_to_idx]

def resize_keep_ratio(img, target_h=IMG_H, target_w=IMG_W):
    img = img.convert("L")
    w, h = img.size

    new_w = int(w * (target_h / h))
    new_w = max(1, min(new_w, target_w))

    img = img.resize((new_w, target_h))

    canvas = Image.new("L", (target_w, target_h), color=255)
    canvas.paste(img, (0, 0))

    arr = np.array(canvas).astype(np.float32) / 255.0
    arr = 1.0 - arr

    return torch.tensor(arr).unsqueeze(0)

class ArabicOCRDataset(Dataset):
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["image_path"])
        x = resize_keep_ratio(img)
        y = torch.tensor(encode_text(row["text"]), dtype=torch.long)
        return x, y, str(row["text"])

def collate_fn(batch):
    images, labels, texts = zip(*batch)
    images = torch.stack(images)
    label_lengths = torch.tensor([len(l) for l in labels], dtype=torch.long)
    labels = torch.cat(labels)
    return images, labels, label_lengths, texts

train_ds = ArabicOCRDataset("/content/ocr_train_clean.csv")
val_ds = ArabicOCRDataset("/content/ocr_val_clean.csv")

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_ds, batch_size=16, shuffle=False, collate_fn=collate_fn)

class CRNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),
        )

        self.rnn = nn.LSTM(
            input_size=512 * 4,
            hidden_size=256,
            num_layers=2,
            bidirectional=True,
            batch_first=True,
            dropout=0.2
        )

        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        feat = self.cnn(x)
        b, c, h, w = feat.size()
        feat = feat.permute(0, 3, 1, 2)
        feat = feat.reshape(b, w, c * h)
        out, _ = self.rnn(feat)
        out = self.fc(out)
        return out

model = CRNN(num_classes).to(device)

x, y, label_lengths, texts = next(iter(train_loader))
with torch.no_grad():
    logits = model(x.to(device))

print("batch:", x.shape)
print("logits:", logits.shape)
print("sample:", texts[0])
```

```python
def decode_ctc(logits):
    pred = logits.argmax(dim=-1).detach().cpu().numpy()
    texts = []

    for seq in pred:
        out = []
        prev = None
        for idx in seq:
            if idx != blank_idx and idx != prev:
                out.append(idx_to_char.get(int(idx), ""))
            prev = idx
        texts.append("".join(out))

    return texts

criterion = nn.CTCLoss(blank=blank_idx, zero_infinity=True)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

def train_one_epoch():
    model.train()
    total_loss = 0

    for images, labels, label_lengths, texts in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        label_lengths = label_lengths.to(device)

        logits = model(images)
        log_probs = F.log_softmax(logits, dim=-1).permute(1, 0, 2)

        input_lengths = torch.full(
            size=(images.size(0),),
            fill_value=logits.size(1),
            dtype=torch.long,
            device=device
        )

        loss = criterion(log_probs, labels, input_lengths, label_lengths)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5)
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)

def preview_predictions():
    model.eval()
    with torch.no_grad():
        images, labels, label_lengths, texts = next(iter(val_loader))
        logits = model(images.to(device))
        preds = decode_ctc(logits)

        for true, pred in list(zip(texts, preds))[:3]:
            print("TRUE:", true)
            print("PRED:", pred)
            print("-" * 30)

for epoch in range(1, 6):
    loss = train_one_epoch()
    print(f"epoch {epoch} loss: {loss:.4f}")
    preview_predictions()

torch.save({
    "model_state": model.state_dict(),
    "chars": chars,
    "img_h": IMG_H,
    "img_w": IMG_W
}, "/content/arabic_crnn_ctc.pth")

print("saved /content/arabic_crnn_ctc.pth")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
import numpy as np
import json

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

IMG_H = 64
IMG_W = 1024

with open("/content/arabic_ocr_vocab_clean.json", "r", encoding="utf-8") as f:
    vocab = json.load(f)

chars = vocab["chars"]
char_to_idx = vocab["char_to_idx"]
idx_to_char = {int(k): v for k, v in vocab["idx_to_char"].items()}

blank_idx = 0
num_classes = len(chars) + 1

def encode_text(text):
    return [char_to_idx[ch] for ch in str(text) if ch in char_to_idx]

def resize_keep_ratio(img, target_h=IMG_H, target_w=IMG_W):
    img = img.convert("L")
    w, h = img.size
    new_w = int(w * (target_h / h))
    new_w = max(1, min(new_w, target_w))
    img = img.resize((new_w, target_h))
    canvas = Image.new("L", (target_w, target_h), color=255)
    canvas.paste(img, (0, 0))
    arr = np.array(canvas).astype(np.float32) / 255.0
    arr = 1.0 - arr
    return torch.tensor(arr).unsqueeze(0)

class ArabicOCRDataset(Dataset):
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["image_path"])
        x = resize_keep_ratio(img)
        y = torch.tensor(encode_text(row["text"]), dtype=torch.long)
        return x, y, str(row["text"])

def collate_fn(batch):
    images, labels, texts = zip(*batch)
    images = torch.stack(images)
    label_lengths = torch.tensor([len(l) for l in labels], dtype=torch.long)
    labels = torch.cat(labels)
    return images, labels, label_lengths, texts

train_ds = ArabicOCRDataset("/content/ocr_train_short.csv")
val_ds = ArabicOCRDataset("/content/ocr_val_short.csv")

train_loader = DataLoader(train_ds, batch_size=8, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_ds, batch_size=8, shuffle=False, collate_fn=collate_fn)

class CRNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),      # W / 2

            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),      # W / 4

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),
        )

        self.rnn = nn.LSTM(
            input_size=512 * 4,
            hidden_size=256,
            num_layers=2,
            bidirectional=True,
            batch_first=True,
            dropout=0.2
        )

        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        feat = self.cnn(x)
        b, c, h, w = feat.size()
        feat = feat.permute(0, 3, 1, 2)
        feat = feat.reshape(b, w, c * h)
        out, _ = self.rnn(feat)
        out = self.fc(out)
        return out

model = CRNN(num_classes).to(device)

x, y, label_lengths, texts = next(iter(train_loader))
with torch.no_grad():
    logits = model(x.to(device))

print("batch:", x.shape)
print("logits:", logits.shape)
print("max label length:", max(label_lengths).item())
print("sample:", texts[0])
```

```python
import os, zipfile, re, json
import pandas as pd
from sklearn.model_selection import train_test_split

DRIVE_DIR = "/content/drive/MyDrive/handwritten"

TRAIN_ZIP = os.path.join(DRIVE_DIR, "train.zip")
GT_XLSX = os.path.join(DRIVE_DIR, "Groundtruth-Unicode.xlsx")

WORK_DIR = "/content/khatt_work"
TRAIN_DIR = os.path.join(WORK_DIR, "train")

os.makedirs(WORK_DIR, exist_ok=True)

if not os.path.exists(TRAIN_DIR):
    with zipfile.ZipFile(TRAIN_ZIP, "r") as z:
        z.extractall(TRAIN_DIR)
    print("unzipped train.zip")

gt = pd.read_excel(GT_XLSX, header=None)
gt = gt.iloc[:, :2]
gt.columns = ["filename", "text"]

image_paths = {}
for root, dirs, files in os.walk(TRAIN_DIR):
    for f in files:
        if f.lower().endswith((".png", ".jpg", ".jpeg", ".tif", ".tiff")):
            image_paths[f] = os.path.join(root, f)

rows = []
for _, row in gt.iterrows():
    filename = row["filename"]
    if filename in image_paths:
        rows.append({
            "image_path": image_paths[filename],
            "text": str(row["text"])
        })

df = pd.DataFrame(rows)
print("matched rows:", len(df))

def clean_arabic_text(text):
    text = str(text)
    text = text.replace("\xa0", " ")
    text = re.sub(r"\s+", " ", text).strip()

    diacritics = "ًٌٍَُِّْـ"
    for ch in diacritics:
        text = text.replace(ch, "")

    allowed = set(" ابتثجحخدذرزسشصضطظعغفقكلمنهويىءآأؤإئة0123456789.,،؛:؟?!()-/\"")
    text = "".join(ch for ch in text if ch in allowed)

    return text.strip()

df["text"] = df["text"].apply(clean_arabic_text)
df = df[df["text"].str.len() > 0].reset_index(drop=True)

train_df, val_df = train_test_split(df, test_size=0.15, random_state=42, shuffle=True)

train_csv = os.path.join(DRIVE_DIR, "ocr_train_clean.csv")
val_csv = os.path.join(DRIVE_DIR, "ocr_val_clean.csv")
vocab_json = os.path.join(DRIVE_DIR, "arabic_ocr_vocab_clean.json")

train_df.to_csv(train_csv, index=False, encoding="utf-8-sig")
val_df.to_csv(val_csv, index=False, encoding="utf-8-sig")

all_text = "".join(train_df["text"].astype(str).tolist() + val_df["text"].astype(str).tolist())
chars = sorted(set(all_text))

char_to_idx = {ch: i + 1 for i, ch in enumerate(chars)}
idx_to_char = {i + 1: ch for i, ch in enumerate(chars)}

vocab = {
    "chars": chars,
    "char_to_idx": char_to_idx,
    "idx_to_char": idx_to_char
}

with open(vocab_json, "w", encoding="utf-8") as f:
    json.dump(vocab, f, ensure_ascii=False, indent=2)

print("saved:", train_csv)
print("saved:", val_csv)
print("saved:", vocab_json)
print("train:", len(train_df))
print("val:", len(val_df))
print("num chars:", len(chars))
```

```python
# Install once
import subprocess, sys
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 
                'transformers', 'torch', 'pillow'], check=False)

from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import torch
from pathlib import Path
import tqdm as notebook_tqdm

# Setup
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

# Load pretrained Arabic handwriting model
model_name = "RayR1/trocr-base-arabic-handwritten"
processor = TrOCRProcessor.from_pretrained(model_name)
trocr_model = VisionEncoderDecoderModel.from_pretrained(model_name).to(device)
print("TrOCR loaded")

def recognize_with_trocr(image_path):
    """Recognize a handwritten Arabic word using TrOCR."""
    img = Image.open(image_path).convert("RGB")
    pixel_values = processor(
        images=img, return_tensors="pt"
    ).pixel_values.to(device)
    
    generated_ids = trocr_model.generate(
        pixel_values, max_length=64
    )
    pred_text = processor.batch_decode(
        generated_ids, skip_special_tokens=True
    )[0]
    
    # Extract individual letters
    letters = [ch for ch in pred_text if ch.strip()]
    return pred_text, letters

# Test on your val images
ROOT = Path(r'F:\newbach\Arabic-Handwritten-Characters-Recognition-using-CNN')
val_images = list((ROOT / 'my_dataset' / 'val' / 'images').glob('*.*'))[:10]

print("\nTesting TrOCR on 10 val images:")
print("=" * 50)
for img_path in val_images:
    pred_text, letters = recognize_with_trocr(img_path)
    print(f"Image: {img_path.name}")
    print(f"  Predicted: {pred_text}")
    print(f"  Letters  : {letters}")
    print()
```
